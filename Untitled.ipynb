```python
import pandas as pd
import requests
from io import StringIO

# GitHub repository URL
github_url = 'https://raw.githubusercontent.com/abhyudaya29/Consignment_Data/main/SCMS_Delivery_History_Dataset.csv'

# Fetch the dataset from GitHub
response = requests.get(github_url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the CSV content
    csv_data = StringIO(response.text)
    
    # Load the CSV into a pandas DataFrame
    df = pd.read_csv(csv_data).fillna(0)
    
    # Now you can work with the DataFrame (df)
    print(df.head())
else:
    print('Failed to fetch the dataset from GitHub')
```

```python

itemcount = df['Country'].value_counts().nlargest(10)
print("Top 10 countries wise count\n")
print(itemcount)

import seaborn as sns
import matplotlib.pyplot as plt

# This line prints the top 10 countries and their respective counts.
# It should correctly display the most frequent countries in your dataset.
# Make sure that 'df' is properly defined and contains the 'Country' column.
# If you see the correct output for this part, your data is being processed correctly.
# If not, check the content and structure of your DataFrame.
sns.set_context("talk", font_scale=1)
plt.figure(figsize=(22,6))
sns.countplot(data=df, x='Country', order=df['Country'].value_counts().nlargest(10).index)
plt.title('Top 10 countries wise')

plt.ylabel('Total Count')
plt.xlabel('Country name')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()  # Adjust layout for better appearance
plt.show()
```

```python
## total pack price for top 15 countries
total_price=df.groupby(['Country'])['Pack Price'].sum().nlargest(15)
print("Total Pack Price for top 10 countries")
print(total_price)
plt.figure(figsize=(22,6))
GraphData=df.groupby(['Country'])['Pack Price'].sum().nlargest(15)
GraphData.plot(kind='bar')
plt.ylabel('Total Pack Price')
plt.xlabel('Country Name')
```
